```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
headers = ['filename','metric1','metric2','metric3','pixels']
sift_AS = pd.read_csv("descriptors/sift_desc/AS_metrics.csv",names=headers,skiprows=1)
sift_RB = pd.read_csv("descriptors/sift_desc/RB_metrics.csv",names=headers,skiprows=1)
sift_RK = pd.read_csv("descriptors/sift_desc/RK_metrics.csv",names=headers,skiprows=1)

surf_AS = pd.read_csv("descriptors/surf_desc/metrics/AS_surf.csv",names=headers,skiprows=1)
surf_RB = pd.read_csv("descriptors/surf_desc/metrics/RB_surf.csv",names=headers,skiprows=1)
surf_RK = pd.read_csv("descriptors/surf_desc/metrics/RK_surf.csv",names=headers,skiprows=1)

orb_AS = pd.read_csv("descriptors/orb_desc/AS_dataset_metrics.csv",names=headers,skiprows=1)
orb_RB = pd.read_csv("descriptors/orb_desc/RB_dataset_metrics.csv",names=headers,skiprows=1)
orb_RK = pd.read_csv("descriptors/orb_desc/RK_dataset_metrics.csv",names=headers,skiprows=1)
```

```python
data_arrays = [(orb_RK, orb_RB, orb_AS), (surf_RK, surf_RB, surf_AS), (sift_RK, sift_RB, sift_AS)]
line_styles = ['solid', 'dashed', 'dotted']
colors = ['red', 'blue', 'green']
names = ['orb', 'surf', 'sift']
legends = ['RK', 'RB', 'AS']
metrics = ['Right features ratio','Average distance','Time']

eps = 1e-2
for i in range(3):
    fig, axs = plt.subplots(1, 3)
    fig.suptitle(metrics[i], fontsize=30)
    fig.set_figheight(10)
    fig.set_figwidth(40)
    
    if i < 2:
        
        for j in range(3):

            for k in range(3):
                #axs[i, j].set_ylim(min(data_arrays[j][k][f'metric{i + 1}']) - eps, 
                #                   max(data_arrays[j][k][f'metric{i + 1}']) + eps)
                axs[j].set_title(names[j], fontsize=20)
                if i == 0:
                    axs[j].set_ylabel('Right features ratio', fontsize=16)
                else:
                    axs[j].set_ylabel('Average distance', fontsize=16)
                
                axs[j].set_xlabel('images', fontsize=16)
#                 axs[j].legend(legends)
                axs[j].plot(data_arrays[j][k][f'metric{i + 1}'], color=colors[k], linestyle=line_styles[k])
    else:
        
         for j in range(3):

            for k in range(3):
                #axs[i, j].set_ylim(min(data_arrays[j][k][f'metric{i + 1}']) - eps, 
                #                   max(data_arrays[j][k][f'metric{i + 1}']) + eps)
                axs[j].set_title(names[j], fontsize=20)
                axs[j].set_xlabel('pixels', fontsize=16)
                axs[j].set_ylabel('time', fontsize=16)
                #axs[j].legend(legends)
                axs[j].scatter(data_arrays[j][k]['pixels'], data_arrays[j][k][f'metric{i + 1}'],
                               color=colors[k], linestyle=line_styles[k])
                
    if i == 0:
        fig.legend(legends)
    
    if i == 1:
        fig.legend(legends)
   
    if i == 2:
        fig.legend(legends)
```
